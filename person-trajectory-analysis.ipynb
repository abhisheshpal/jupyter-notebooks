```python
! pip install -U pymongo==2.8
from pymongo import MongoClient
from csv import DictWriter
from pprint import pprint

client = MongoClient('mongodb://10.5.42.18:62345')
db = client.message_store_y4
coll_traj = db.trajectory_person
coll_people = db.people_perception

def export_traj(entry, dw):
        d = {
            'robot_x': entry['robot']['position']['x'],
            'robot_y': entry['robot']['position']['y'],
            'waypoint': entry['closest_node'],
            'timestamp': entry['header']['stamp']['secs']*1000 + int(entry['header']['stamp']['nsecs']/1000000)
        }
        uuids = entry['uuids']
        poses = entry['people_tracker']['poses']
        velocities = entry['people_tracker']['velocities']
        n = len(uuids)
        for i in range(0,n):
            d['uuid'] = uuids[i]
            #d['timestamp'] = 
            d['pose_x'] = poses[i]['position']['x']
            d['pose_y'] = poses[i]['position']['y']
            d['vel_x'] = velocities[i]['x']
            d['vel_y'] = velocities[i]['y']
        dw.writerow(d)
        #pprint(d)

def export(coll_people, filename):
    with open(filename, 'w') as csvfile:
        fieldnames = ['timestamp', 'uuid',
                      'robot_x', 'robot_y',
                      'pose_x', 'pose_y',
                      'vel_x', 'vel_y',
                      'waypoint']
        dw = DictWriter(csvfile, fieldnames=fieldnames, extrasaction='ignore')
        dw.writeheader()
        for e in coll_people.find():
            export_traj(e, dw)

export(coll_people, 'out.csv')
#pprint(coll_people.find_one())
```
